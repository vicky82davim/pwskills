```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
# Independent and dependent features
X=df.iloc[:,:-1]
y = dataset.target
```

```python
X,y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
# In decision tree no need to do standardization, feature selection, normalization as no need to make 
# smaller values and calculation
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
## Post pruning with max_depth=2
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
## prediction
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
print(classification_report(y_pred, y_test))
```

```python
## Prepruning and hyper parameter tuning for huge data
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier = DecisionTreeClassifier()
clf = GridSearchCV(treeclassifier, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
## Train the data
clf.fit(X_train, y_train)
```

```python
## below parametrs will be used as prepruning
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_pred, y_test)
score
```

```python
print(classification_report(y_pred, y_test))
```
