```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
"""
"""
scatter plot visualize the data in terms of dots over graph.
"""
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is title")
plt.show()
```

```python
"""
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
"""

"""
bar plot is used to show the data in bar format
"""
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company,profit)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is title")
plt.show()
```

```python
"""
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
"""

"""
A box plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, 
and maximum. In a box plot, we draw a box from the first quartile to the third quartile. 
A vertical line goes through the box at the median.
"""
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
_data = [box1, box2] 
```

```python
_fig = plt.figure(figsize =(8, 5)) 
_ax = _fig.add_axes([0, 0, 1, 1]) 
_bp = _ax.boxplot(_data) 
plt.show() 
```
