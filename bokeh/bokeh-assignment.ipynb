```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_file
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
output_file('flower.html')
p = figure(title='Flower')
p.xaxis.axis_label = "X Label"
p.yaxis.axis_label = "Y Label"
p.circle(flowers['petal_width'], flowers['petal_length'])
show(p)
```

```python
"""
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

"""
"""
Any plot is usually made up of one or many geometrical shapes such as line, circle, rectangle, etc. 
These shapes have visual information about the corresponding set of data. 
In Bokeh terminology, these geometrical shapes are called gylphs.
"""
from bokeh.plotting import figure, output_file, show
output_file("plot1.html")
p = figure(width=400, height=400)
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)
show(p)
```

```python
"""
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

"""
x = [2,3,4,5,6]
y=[3,4,5,6,7]
output_file('cust1.html')
p = figure(title='line grapg')
p.xaxis.axis_label = "Petal length"
p.yaxis.axis_label = "Petal width"
p.scatter(x,y, fill_color='red', legend_label = 'red', size=20)
show(p)
```
